# Analysis with Pandas and Numpy

## Importing Libraries and Creating Dictionaries for Analysis

```python
import numpy as np
import pandas as pd
# Naming the store to fit the dataframe
General_Store = {
    'product _id':[23,96,97,15,87],
    'name':['computer','python workout','pandas workout', 'banana','sandwich'],
    'wholesale_price':[500,35,35,0.5,3.0],
    'retail_price': [1000,75,75,1,5],
    'sales': [100,1000,500,200,300]
}
df = pd.DataFrame(General_Store)
df
```

## Finding Net Sales 

```python
#Calculating the Net Revenue per Product

df["net_revenue_per_product"] = (df["retail_price"] - df["wholesale_price"]) * df["sales"]
df
```

```python
# Calculating the Total Revenue Received from Sales
Total_Revenue = df["net_revenue_per_product"].sum()
print("The Total Revenue received from sales is " + str(Total_Revenue))
```

```python
# Finding The product whose retail price is twice more than the wholesale price
df.loc[(df["retail_price"] >(df["wholesale_price"]*2))]
```

## Comparative Revenues between products

```python
# Finding how much General Store made from  Food vs Computers vs Books

food = df.loc[[3,4], 'net_revenue_per_product'].sum()
print('Total amount made from food is -', food)
books = df.loc[[2,1], 'net_revenue_per_product'].sum()
print('Total amount made from books is -', books)
computer = df.loc[[0], 'net_revenue_per_product'].sum()
print('Total amount made from computer is -', computer)
```

## Discounted Net Sales 

```python
 # Caluculating the New Net Revenue after 30% ndiscount negotiation on Wholesale Price

df['new_wholesale_price'] = df['wholesale_price'] - 0.3 * df['wholesale_price']
df
```

```python
df['new_net_revenue_per_ product'] = (df['retail_price'] - df['new_wholesale_price']) * df['sales']
df
```
